The computer science perspective is that vectors are ordered lists of numbers. For example, if you were doing some analytics about house prices, and the only features you cared about were square footage and price, you might model each house as a pair of numbers, the first indicating square footage, and the second indicating price.

This notebook explores basic vector operations and visualizations using NumPy and Matplotlib.

- Vectors are introduced as ordered lists of numbers, like representing house features.
- We demonstrated vector addition and subtraction.
- We visualized 2D vectors and adjusted plot limits, added a grid, legend, and title.
- We also visualized a 3D vector by plotting a line from the origin.
- Finally, we showed how to perform scalar multiplication (scaling) on vectors, including negative scaling.

```python
import numpy as np
```

```python
# performing vector operations like addition and substraction
b = np.array([300,200])
c = np.array([400,500])
print(b+c)
print(b-c)
```

```python
# visualizing 2d space
import numpy as np
import matplotlib.pyplot as plt

# define vectors
v1 = np.array([3, 2])
v2 = np.array([1, 4])

# create a plot
plt.figure(figsize=(5, 5))
plt.axhline(0, color='gray', linewidth=0.5)  # x-axis
plt.axvline(0, color='gray', linewidth=0.5)  # y-axis

# plot the vector using quiver
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector 1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector 2')
```

```python
# visualizing 2d space
import numpy as np
import matplotlib.pyplot as plt

# define vectors
v1 = np.array([3, 2])
v2 = np.array([1, 4])

# create a plot
plt.figure(figsize=(5, 5))
plt.axhline(0, color='gray', linewidth=0.5)  # x-axis
plt.axvline(0, color='gray', linewidth=0.5)  # y-axis

# plot the vector using quiver
plt.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='Vector 1')
plt.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label='Vector 2')

# adjusted limit
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid() # provides a grid
plt.legend() # provides the label at the right hand side
plt.title("Vector Visualization") # it's the title
plt.show() # it's used to display stuff
```

```python
# visualizing a vector in 3d space
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# define the vector
v = np.array([2, 3, 4])

# create a 3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# plot the vector as a line from the origin to (x,y,z)
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b')
```

```python
# visualizing a vector in 3d space
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# define the vector
v = np.array([2, 3, 4])

# create a 3d plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# plot the vector as a line from the origin to (x,y,z)
ax.quiver(0, 0, 0, v[0], v[1], v[2], color='b')

# set labels and also setting the limit
ax.set_xlim([-3, 5])
ax.set_ylim([-3, 5])
ax.set_zlim([-3, 5])
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```
